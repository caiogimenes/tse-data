```python
import pandas as pd
```

```python
data = pd.read_csv('resultados_secao_absolutos/EDUARDO_DA_COSTA_PAES.csv', index_col=0)
```

```python
data.loc[data['votos_recebidos'] > data['eleitores_totais']]
```

```python
import tse
```

```python
database = 'databases/eleicoes_2024.sql'
analise = tse.TseAnalysis(
    database={
        'path': database,
        'alias': 'geral'
    })
```

```python
analise.attach_database([])
```

```python
query = """
    SELECT
        NR_LOCAL_VOTACAO AS local,
        NR_ZONA AS zona,
        NR_SECAO AS secao,
        NR_LATITUDE AS latitude,
        NR_LONGITUDE AS longitude,
        QT_ELEITOR_SECAO as eleitores
    FROM
        geral.locais
    WHERE
        CD_MUNICIPIO = 60011
"""
```

```python
locais = pd.read_sql(query, analise.conn)
```

```python
locais.head()
```

```python
data_locais = pd.merge(
    data,
    locais,
    how='inner',
    on=['local', 'zona', 'secao']
)
data_locais.head()
```

```python
inconsis = data_locais.loc[data['votos_recebidos'] > data['eleitores_totais']]
```

```python
inconsis
```

```python
inconsis.groupby('local').agg(
    {
        'votos_recebidos': 'sum',
        'eleitores' : 'sum'
    }
)
```

```python
import folium
```

```python
map = folium.Map(location=[-22.908333, -43.196388])
```

```python
for i, row in inconsis.iterrows():
    folium.Marker(location=[row['latitude'], row['longitude']]).add_to(map)
```
